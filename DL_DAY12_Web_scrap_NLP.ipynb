```
# Reading from a File
f = open('mywebdata.txt')
mydata = f.read()
f.close()
```

```
# Splitting the data by new line
mydata=mydata.split()
```

```
mydata
```

```
# NLTK graphs
import matplotlib.pyplot as plt
import nltk

%matplotlib inline    # incase matplotlib is not supporting by default with pandas or nltk
# to replace plt.show() from any platform
```

```
[i for i in dir(nltk) if 'Freq' in i] # to plot
```

```
nlpdata=nltk.FreqDist(mydata)
```

```
nlpdata.plot(10)
```

```
# we can remove stopwords from nlpdata and then plot the graph
from nltk.corpus import stopwords
```

```
# removing stop words
removedata=[i for i in mydata if i.lower() not in stopwords.words('english')]
```

```
# applying NLTK FreqDist
nlpdata1 = nltk.FreqDist(removedata)
```

```
nlpdata1.plot(20)
```
